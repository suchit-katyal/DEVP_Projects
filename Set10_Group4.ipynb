<h2> ANALYSING THE RELATIONSHIP B/W NO. OF ROOMS AND PRICES </h2>

```python
sns.barplot(x='Rooms',y='Price',data=df)
# 7 rooms apartment are having the highest prices.
```

```python
sns.countplot(x='Method',data=df)
# To find out the category of property
```

```python
# Most of the properties are SOLD and least were SOLD AFTER Auction
```

```python
sns.barplot(x=df['Price'])
```

```python
sns.barplot(x='Type',y='Landsize',data=df)
# To find out the landsize and house type relationship, house has the max landsize and Townhouse the least
```

```python
sns.barplot(x='Type',y='Landsize',data=df, hue='Car')
# To find out the relationship among landsize and Type of apartment with respect to the no. of cars parking included
```

```python
sns.barplot(x='Type',y='Landsize',data=df, hue='Bathroom')
# To find out the relationship among landsize and Type of apartment with respect to the no. of Bathrooms
```

```python
plt.figure(figsize=(90,9))
sns.barplot(x='Distance',y='Price',data=df)
#To know the pricing policy wrt to the distance from the centre. 9 and 3 distance metrics are the most expensive
```

```python
sns.boxplot(x='Price',y='Landsize',data=df)
```

```python

df4=df.pivot_table(index='Type',columns='Regionname',values='YearBuilt') 
sns.heatmap(df4)
# To find the yearwise built apartments in different regions wrt to apartment type.
```

```python
plt.figure(figsize=(20,9))
plt.scatter(x='Regionname',y='Price',data=df) 
plt.show()
# To find the regionwise pricing
```

```python

df7=df.pivot_table(index='Rooms',columns='Car',values='Bathroom') 
sns.heatmap(df7,linewidth=1,cmap="Blues")
# To find the apartment orientation with rooms, bathrooms and car parking
```

```python
plt.figure(figsize=(20,6)) 
_=sns.violinplot(x="Regionname", y="Price", data=df)
# To find region wise average pricing
```

```python
sns.barplot(x='Type',y='Price',data=df)
# Apartment type and pricing
```

```python
plt.figure(figsize=(200,60))
sns.barplot(x='YearBuilt',y='Price',data=df)
# Old apartments are expensive
```

```python
# To analyze price with respect to region and Method(Property Status)
df9=df.pivot_table(index='Regionname',columns='Method',values='Price') 
sns.heatmap(df9)
```

```python
# Apartments with 9 car parks are the most expensive and with one car park the least
sns.barplot(x='Car',y='Price',data=df)
```
