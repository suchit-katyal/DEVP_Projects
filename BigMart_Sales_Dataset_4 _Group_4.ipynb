```python
# Finding Correlation,
corrMatrix=df.corr()
corrMatrix
```

```python
# It is the two-dimensional graphical representation of the data where the individual values that are contained in the matrix are represented as colours.
sns.heatmap(corrMatrix, annot=True)
plt.show()
# outlet sales show positive correlation with marked price but for rest of the data we could not see any strong correlation between variables.
```

```python
# Crosstab between the location and the size of the outlet
pd.crosstab(df.Outlet_Location_Type,df.Outlet_Size)
```

```python
# Barplot to determine which type of item is most expensive?
plt.figure(figsize=(20,10))
a = sns.barplot(x="Item_Type",y="Item_MRP",data=df)
a.set(xlabel="Type of Items",ylabel="Price of the Item",title="Most expensive Item")
```

```python
# Which year has maximum numer of outlets established?
plt.figure(figsize=(10,10))
sns.countplot(x="Outlet_Establishment_Year", data=df)
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(y='Item_Outlet_Sales',x='Item_MRP',data=df)
plt.xlabel('Item MRP')
plt.ylabel('Item Outlet Sales')
#Items with higher MRP are sold maximum
```

```python
#scatter plot 
#Where the visiblity of Items is Zero, which suggest those items kept behind in shelf and amlost have no visiblity can also be sold. This show Consumer tend to search for their own products
plt.figure(figsize=(10,5))
plt.scatter('Item_Visibility','Item_Outlet_Sales',data=df)
plt.xlabel('Item_Visibility')
plt.ylabel('Item_Outlet_Sales')
```
